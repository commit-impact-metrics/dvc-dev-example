*** Обоснование весов и критерии для каждой метрики ***

**1. Объем данных (data_volume)**

Вес: 0.2\
Почему такой вес: Большой объем данных увеличивает вероятность технических ошибок, требует масштабируемости и тщательного тестирования.\
- Критерии оценки (1-5):\
1 - Малый объем данных (до 100K записей).\
2 - Средний объем данных (100K–500K записей).\
3 - Большой объем данных (500K–1M записей).\
4 - Очень большой объем (1M–10M записей).\
5 - Экстремально большой объем (более 10M записей).

**2. Критичность процесса (criticality)**

Вес: 0.25\
Почему такой вес: Наиболее значимая метрика, так как критичность процесса определяет возможные бизнес-убытки при сбоях.\
- Критерии оценки (1-5):\
1 - Некритичные данные, не влияющие на основные процессы.\
2 - Второстепенные данные, поддерживающие вспомогательные процессы.\
3 - Умеренная критичность, частичное влияние на ключевые процессы.\
4 - Высокая критичность, данные влияют на основные процессы.\
5 - Критичные данные для жизненно важных процессов (например, финансы, безопасность).

**3. Сложность реализации (complexity)**

Вес: 0.15\
Почему такой вес: Сложные реализации повышают вероятность технических ошибок и требуют более тщательного тестирования.\
- Критерии оценки (1-5):\
1 - Простые изменения, минимальная сложность.\
2 - Умеренная сложность, стандартные ETL-процессы.\
3 - Средняя сложность, уникальные трансформации.\
4 - Высокая сложность, значительная переработка системы.\
5 - Очень высокая сложность, внедрение нового технологического стека.

**4. Частота использования данных (data_frequency)**

Вес: 0.15\
Почему такой вес: Часто используемые данные требуют повышенного внимания к их качеству и доступности.\
- Критерии оценки (1-5):\
1 - Редкое использование (раз в месяц).\
2 - Умеренное использование (раз в неделю).\
3 - Частое использование (ежедневно).\
4 - Очень частое использование (почасовое).\
5 - Использование в реальном времени.

**5. Чувствительность к ошибкам (error_sensitivity)**

Вес: 0.2\
Почему такой вес: Ошибки в чувствительных данных могут привести к крупным финансовым или репутационным потерям.\
- Критерии оценки (1-5):\
1 - Минимальная чувствительность, ошибки некритичны.\
2 - Умеренная чувствительность, влияние на незначительные процессы.\
3 - Средняя чувствительность, влияние на бизнес-отчеты.\
4 - Высокая чувствительность, ошибки заметны пользователям.\
5 - Критическая чувствительность, ошибки недопустимы.

**6. Сложность интеграции (integration_complexity)**

Вес: 0.1\
Почему такой вес: Чем сложнее интеграция, тем выше вероятность ошибок, но обычно это решается архитектурными подходами.\
- Критерии оценки (1-5):\
1 - Нет интеграций.\
2 - Простая интеграция (1-2 системы).\
3 - Умеренная интеграция (3-5 систем).\
4 - Сложная интеграция (5-10 систем).\
5 - Очень сложная интеграция (более 10 систем).

**7. Срочность внедрения (urgency)**

Вес: 0.05\
Почему такой вес: Срочность повышает вероятность пропуска ошибок, но её влияние можно уменьшить планированием.\
- Критерии оценки (1-5):\
1 - Внедрение без дедлайна.\
2 - Низкий приоритет, срок в течение месяца.\
3 - Средний приоритет, срок в течение недели.\
4 - Высокий приоритет, срок в течение 3 дней.\
5 - Крайне срочное внедрение.

**8. Опыт команды (team_experience)**

Вес: 0.1\
Почему такой вес: Опыт команды снижает вероятность ошибок.\
- Критерии оценки (1-5):\
1 - Новая команда, нет опыта.\
2 - Минимальный опыт работы.\
3 - Средний уровень опыта.\
5 - Высокий уровень опыта.\
6 - Экспертная команда.

**9. История прошлых ошибок (past_errors)**

Вес: 0.1\
Почему такой вес: История ошибок позволяет прогнозировать вероятность проблем.\
- Критерии оценки (1-5):\
1 - Ошибки отсутствуют.\
2 - Единичные ошибки.\
3 - Умеренное количество ошибок.\
4 - Частые ошибки.\
5 - Постоянные и критические ошибки.

Суммарный вес: 0.2 + 0.25 + 0.15 + 0.15 + 0.2 + 0.1 + 0.05 + 0.1 + 0.1=1.0.

```python
# Определение метрик и их весов
metrics = {
    "data_volume": {"weight": 0.2, "description": "Объем данных"},
    "criticality": {"weight": 0.25, "description": "Критичность процесса"},
    "complexity": {"weight": 0.15, "description": "Сложность реализации"},
    "data_frequency": {"weight": 0.15, "description": "Частота использования данных"},
    "error_sensitivity": {"weight": 0.2, "description": "Чувствительность к ошибкам"},
    "integration_complexity": {"weight": 0.1, "description": "Сложность интеграции с другими системами"},
    "urgency": {"weight": 0.05, "description": "Срочность внедрения релиза"},
    "team_experience": {"weight": 0.1, "description": "Опыт команды разработчиков"},
    "past_errors": {"weight": 0.1, "description": "История прошлых ошибок"}
}

# Ввод данных пользователя
user_input = {}
print("Введите оценки для каждой метрики от 1 до 5:")
for metric, details in metrics.items():
    while True:
        try:
            value = int(input(f"{details['description']} (1-5): "))
            if 1 <= value <= 5:
                user_input[metric] = value
                break
            else:
                print("Ошибка: введите значение от 1 до 5.")
        except ValueError:
            print("Ошибка: введите целое число от 1 до 5.")

# Расчет риска
total_risk = sum(metrics[metric]["weight"] * user_input[metric] for metric in metrics)

# Вывод стратегии тестирования
def testing_strategy(level):
    strategies = {
        "Низкий": "Минимальное тестирование ключевых функций.",
        "Средний": "Регрессионное и ручное тестирование.",
        "Высокий": "Полное тестирование + экспертная проверка.",
        "Высокий+": "Полный аудит и автоматизированное тестирование."
    }
    return strategies.get(level, "Пересмотреть подход.")

# Вывод результата
if total_risk < 2:
    risk_level = "Низкий"
elif 2 <= total_risk < 3.5:
    risk_level = "Средний"
elif 3.5 <= total_risk < 4.5:
    risk_level = "Высокий"
else:
    risk_level = "Критический"

print(f"\nОбщий риск: {total_risk:.2f}")
print(f"Уровень риска: {risk_level}")
print(f"Рекомендуемая стратегия тестирования: {testing_strategy(risk_level)}")
```
